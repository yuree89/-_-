```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

##1. 브랜드 표현력(N파워링크/SNS 게시글 수)

```python
# 각 항목의 점수와 이름을 딕셔너리로 정의
data = {
    'Hackers': (50, 62),
    'Eduwill': (27, 23),
    'WowPass': (18, 6),
    'Inflearn': (2, 3),
    'FastCampus': (2, 6)
}

# 딕셔너리로부터 x, y, names 얻기
x = np.array([v[0] for v in data.values()])
y = np.array([v[1] for v in data.values()])
names = list(data.keys())

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([18, 6])
distances = np.sqrt(np.sum((np.vstack((x, y)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
import matplotlib.pyplot as plt

# 각 항목별 점수를 리스트로 정의
x = [50, 27, 18, 2, 2]
y = [62, 23, 6, 6, 3]

# 각 항목별 색상을 리스트로 정의
colors = ['blue', 'yellow', 'purple', 'green', 'red']

# 산점도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y, c=colors)

# 각 점의 이름 표시
labels = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]))

# x축과 y축에 레이블 추가
plt.xlabel('Naver Power Link')
plt.ylabel('Number of Posts')

# 그래프 보이기
plt.show()
```

##3. 소비자: 브랜드 인지도 (N데이터랩/SNS팔로워(N,I,Y)

```python
# 각 항목의 점수와 이름을 딕셔너리로 정의
data = {
    'Hackers': (23, 58),
    'Eduwill': (59, 36),
    'WowPass': (2, 1),
    'Inflearn': (6, 2),
    'FastCampus': (9, 3)
}

# 딕셔너리로부터 x, y, names 얻기
x = np.array([v[0] for v in data.values()])
y = np.array([v[1] for v in data.values()])
names = list(data.keys())

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([2, 1])
distances = np.sqrt(np.sum((np.vstack((x, y)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
import matplotlib.pyplot as plt

# 각 항목별 점수를 리스트로 정의
x = [23, 59, 2, 6, 9]
y = [58, 36, 1, 2, 3]

# 각 항목별 색상을 리스트로 정의
colors = ['blue', 'yellow', 'purple', 'green', 'red']

# 산점도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y, c=colors)

# 각 점의 이름 표시
labels = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]))

# x축과 y축에 레이블 추가
plt.xlabel('Naver DataLab')
plt.ylabel('SNS Follower(N,I,Y)')

# 그래프 보이기
plt.show()
```
