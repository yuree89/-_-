```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

##효율성 확인

```python
# 각 항목의 점수와 이름을 딕셔너리로 정의
data = {
    'Hackers': (0.179, 0.889),
    'Eduwill': (1, 1),
    'WowPass': (0.101, 0.109),
    'Inflearn': (0.719, 0.713),
    'FastCampus': (1, 0.146)
}

# 딕셔너리로부터 x, y, names 얻기
x = np.array([v[0] for v in data.values()])
y = np.array([v[1] for v in data.values()])
names = list(data.keys())

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([0.101, 0.109])
distances = np.sqrt(np.sum((np.vstack((x, y)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
import matplotlib.pyplot as plt

# 각 항목별 점수를 리스트로 정의
x = [0.179, 1, 0.101, 0.719, 1]
y = [0.889, 1, 0.109, 0.713, 0.146]

# 각 항목별 색상을 리스트로 정의
colors = ['blue', 'yellow', 'purple', 'green', 'red']

# 산점도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y, c=colors)

# 각 점의 이름 표시
labels = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]))

# x축과 y축에 레이블 추가
plt.xlabel('Search Volume by Channel')
plt.ylabel('SNS Awareness')

# 그래프 보이기
plt.show()
```
