##결제 등록시간 분포

```python
df2 = pd.read_excel("/content/drive/MyDrive/Ubion/와우패스 데이터/6 2020-10-01 _ 2022-09-30 결제 데이터_230330.xlsx")
df2
```

```python
df2['결제시간'].value_counts()
```

```python
plt.hist(df2['결제시간'], bins=24, alpha=0.5, color='steelblue', edgecolor='black')
plt.xlabel('결제시간')
plt.ylabel('빈도')
plt.title('결제 등록시간 분포')
plt.xticks(range(0, 24))
plt.show()
```

```python
df2
```

```python
df2['요일'] = df2['결제일'].dt.day_name()
```

```python
df2['요일'].value_counts()
```

```python
# 요일별 데이터 개수(count) 계산
요일별_데이터 = df2['요일'].value_counts()

# 막대 그래프에 다른 색상 지정
colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink']  # 각 요일에 대한 색상 리스트
plt.bar(요일별_데이터.index, 요일별_데이터.values, color=colors)

plt.xlabel('요일')
plt.ylabel('데이터 개수')
plt.title('요일별 데이터 분포')
plt.show()
```
