```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

##1. 브랜드(표현력/이미지)

```python
# 각 항목의 점수와 이름을 딕셔너리로 정의
data = {
    'Hackers': (5, 5),
    'Eduwill': (4, 4),
    'WowPass': (3, 3),
    'Inflearn': (1, 2),
    'FastCampus': (2, 1)
}

# 딕셔너리로부터 x, y, names 얻기
x = np.array([v[0] for v in data.values()])
y = np.array([v[1] for v in data.values()])
names = list(data.keys())

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([3, 3])
distances = np.sqrt(np.sum((np.vstack((x, y)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
import matplotlib.pyplot as plt

# 각 항목별 점수를 리스트로 정의
x = [5, 4, 3, 2, 1]
y = [5, 4, 3, 1, 2]

# 각 항목별 색상을 리스트로 정의
colors = ['blue', 'yellow', 'purple', 'green', 'red']

# 산점도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y, c=colors)

# 각 점의 이름 표시
labels = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]))

# x축과 y축에 레이블 추가
plt.xlabel('Brand Expression')
plt.ylabel('Brand Image')

# 그래프 보이기
plt.show()
```

##2-1. 자격증 시장에서의 상품(서비스 수준/제품의 다양성/가격)

```python
import numpy as np

# 각 항목들의 점수 초기화
data = {
    'Hackers': (5, 4, 2),
    'Eduwill': (3, 5, 1),
    'WowPass': (4, 3, 3),
    'Inflearn': (2, 2, 5),
    'FastCampus': (1, 1, 4)
}
names = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([data['WowPass']])
scores = np.array(list(data.values()))
distances = np.sqrt(np.sum((scores - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = [5, 3, 4, 2, 1]
y = [4, 5, 3, 2, 1]
z = [2, 1, 3, 5, 4]
names = ['Hackers', 'Eduwill', 'WowPass', 'Inflearn', 'FastCampus']
colors = ['blue', 'yellow', 'purple', 'green', 'red']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(names)):
    ax.scatter(x[i], y[i], z[i], s=100, c=colors[i])
    ax.text(x[i], y[i], z[i], names[i], fontsize=12)

ax.set_xlabel('Service')
ax.set_ylabel('Product(Certificate)')
ax.set_zlabel('Price')

plt.show()
```
