```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

##1. 브랜드(표현력/이미지)

```python
# 각 항목들의 점수 초기화
x = np.array([5, 4, 3, 2, 1])
y = np.array([5, 4, 3, 1, 2])
names = ['Hackers', 'Eduwill', 'WowPass', 'FastCampus', 'Inflearn']

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([3, 3])
distances = np.sqrt(np.sum((np.vstack((x, y)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
import matplotlib.pyplot as plt

# 각 항목별 점수를 리스트로 정의
x = [5, 4, 3, 2, 1]
y = [5, 4, 3, 1, 2]

# 각 항목별 색상을 리스트로 정의
colors = ['blue', 'yellow', 'purple', 'red', 'green']

# 산점도 그리기
fig, ax = plt.subplots()
ax.scatter(x, y, c=colors)

# 각 점의 이름 표시
labels = ['Hackers', 'Eduwill', 'WowPass', 'FastCampus', 'Inflearn']
for i, label in enumerate(labels):
    ax.annotate(label, (x[i], y[i]))

# 그래프 보이기
plt.show()
```

##2-1. 자격증 시장에서의 상품(서비스 수준/제품의 다양성/가격)

```python
import numpy as np

# 각 항목들의 점수 초기화
x = np.array([5, 3, 4, 1, 2])
y = np.array([4, 5, 3, 1, 2])
z = np.array([4, 5, 3, 2, 1])
names = ['Hackers', 'Eduwill', 'WowPass', 'FastCampus', 'Inflearn']

# 와우패스와 각 항목들 간의 유클리디안 거리 계산
wowpass_scores = np.array([z[names.index('WowPass')], y[names.index('WowPass')], x[names.index('WowPass')]])
distances = np.sqrt(np.sum((np.vstack((z, y, x)).T - wowpass_scores) ** 2, axis=1))

# 결과 출력
for i, name in enumerate(names):
    print(f'{name}와 WowPass 간의 거리: {distances[i]:.2f}')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = [5, 3, 4, 1, 2]
y = [4, 5, 3, 1, 2]
z = [4, 5, 3, 2, 1]
names = ['Hackers', 'Eduwill', 'WowPass', 'FastCampus', 'Inflearn']
colors = ['blue', 'yellow', 'purple', 'red', 'green']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(names)):
    ax.scatter(x[i], y[i], z[i], s=100, c=colors[i])
    ax.text(x[i], y[i], z[i], names[i], fontsize=12)

ax.set_xlabel('Service')
ax.set_ylabel('Product(Certificate)')
ax.set_zlabel('Price')

plt.show()
```
